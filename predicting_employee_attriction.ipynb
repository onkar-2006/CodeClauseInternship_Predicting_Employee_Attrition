apply the feature engineering  or the data preprocessing on the dataset
like


1.   transform the catogorical data into numeric

3.   standard the dataset
4.   spliting into train_set and test_set





```python
X =data.drop(columns=["sales"])
y=data["left"]
data["sales"].nunique()
data["sales"].value_counts()
data["salary"].value_counts()
data.rename(columns={"sales":"left"})
```

```python

from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
X["salary"]=le.fit_transform(X["salary"])

```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train ,y_test = train_test_split(X,y ,test_size =0.2 ,random_state=0)

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

for the clsssification use a artifical neuron network

```python
import tensorflow as tf
from tensorflow.keras import models ,Sequential,layers

```

```python
ann=tf.keras.l=models.Sequential()

import tensorflow as tf
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense( units =6 ,activation='relu'))
ann.add(tf.keras.layers.Dense(units=6  ,activation='relu'))

ann.add(tf.keras.layers.Dense( units =1 ,activation='sigmoid'))
```

```python

ann.compile(optimizer="adam" , loss="BinaryCrossentropy" , metrics=["accuracy"])
```

```python

ann.fit(X_train ,y_train ,epochs=20)
```

analyze the result

```python
import numpy as np

y_pred = ann.predict(X_test)

y_pred = (y_pred > 0.5)

y_test_np = y_test.to_numpy()

result = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test_np.reshape(len(y_test_np), 1)), axis=1)

print(result)

```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)

print(ann.predict(sc.transform([[0.38,	0.53,2,157	,3, 0,	1,	0	,1]])) > 0.5)
```
